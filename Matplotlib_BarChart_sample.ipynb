```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
#plot      # X axis Categories Y- axis Frequencies
plt.bar(classes,class1_students)
```

```python
plt.bar(classes,class2_students)
```

```python
plt.bar(classes,class3_students)
```

```python
#code here explore the parameter
plt.bar(classes, class1_students, width = 0.8)                                                          #bydefault it is center   align : {'center', 'edge'}
                                                                #Look at the starting point of the barchart
```

```python
#change size and color  #can change the color
plt.bar(classes, class1_students, width = 0.6, align = "edge") #bydefault it is center   align : {'center', 'edge'}
                                                                #Look at the starting point of the barchart
```

```python
  #edge color,edge line
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "c")   
```

```python
 #Change the line width 
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y",   #edge color
       edgecolor = "k")                               
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "c", #Change the line width 
       edgecolor = "k", linewidth =8) 
                                                 #brightness               #alpha tells you the opacity of the color
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "center", color = "k", #Change the line width 
       edgecolor = "r", linewidth =3)
```

```python
plt.bar(classes, class1_students,width = 0.2, align = "center", color = "k", #Change the line width 
       edgecolor = "r", linewidth =3)
                                                                                #change the line style
```

```python
#continue coding
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y",
       edgecolor = "m", linewidth = 5, alpha = 0.5)  #brightness#alpha tells you the opacity of the color
```

```python
plt.bar(classes, class1_students, width = 0.8, align = "edge", color = "c",
       edgecolor = "r", linewidth = 5, alpha = 0.9, linestyle = ":")   #change the line style
```

```python
#bar gets hidden
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y",
       edgecolor = "m", linewidth = 5, alpha = 0.9, linestyle = "--",
       label =" Class 1 Students",visible = False)    
```

```python
#Change the background Once you use the style.use you dont need to use it again and again
style.use("ggplot")           #Change the background Once you use the style.use you dont need to use it again and again
plt.figure(figsize=(10,7))
plt.bar(classes, class1_students, width = 0.6, align = "edge", color = "c",
       edgecolor = "y", linewidth = 2, alpha = 0.9, linestyle = "--",
       label =" Class 1 Students") #visible=False
```

```python
plt.figure(figsize=(10,7))
plt.bar(classes, class1_students,align = "edge", color = "y",    #we canot use the barwidth in barh method
       edgecolor = "c", linewidth = 2, alpha = 0.9, linestyle = "--",
       label =" Class 1 Students")
```

```python
#name and label everything  also the figure size

plt.figure(figsize=(9,6))
plt.bar(classes, class1_students, width = 0.6, align = "edge", color = "g",
       edgecolor = "r", linewidth = 5, alpha = 0.9, linestyle = "--",
       label =" Class 1 Students") #visible=False

plt.title("Bar Chart of AI Class", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.show()
```
