```python
import matplotlib.pyplot as plt # For visualization
import pandas as pd # for data cleaing & analysis
```

```python
# Dataset taken from kaggel.com
df = pd.read_csv("googleplaystore.csv",nrows = 1000)
df.shape # Get shape of dataset in rows and colms
```

```python
x = df["Rating"]
y = df["Reviews"]
```

```python
plt.scatter(x,y) # Draw scatter plot with parameter x and y
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9)) # figure size ratio 16:9
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha = 0.5,linewidths=0.3,
           edgecolors="g" )
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```
