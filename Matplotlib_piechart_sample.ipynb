```python
#use all the parameters
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
explode = [0.03,0,0.1,0,0]
colors= ["c", 'b','r','y','g']
textprops = {"fontsize":15}

plt.pie(class1_students, 
        labels = classes, 
        explode = explode, 
        colors =colors, 
        autopct = "%0.2f%%", 
        shadow = True, 
        radius = 1.4,
       startangle = 270, 
        normalize=True,
        frame=False,
        textprops =textprops)
plt.show()
```

```python
explode = [0.03,0,0.1,0,0]
colors = ["c", 'b','r','y','g']
textprops = {"fontsize":15}

plt.pie(class1_students, labels = classes, explode = explode, 
        colors =colors, autopct = "%0.2f%%", shadow = True, radius = 2,
       startangle = 270, textprops =textprops)

plt.legend(title = "coding Languages")
plt.show()
```
