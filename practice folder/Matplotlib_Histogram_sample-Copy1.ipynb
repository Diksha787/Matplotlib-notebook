# Ploting Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age = np.random.randint(18,45, (100))   # range between 18 and 45 total number of students is 100
py_students_age = np.random.randint(15,40, (100))
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
plt.hist(ml_students_age)
plt.show()
```

```python
#very importatn to label a histogram only possible if know what nature of the data histogram takes
```

```python
plt.hist(ml_students_age)    #histogram function  

plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

```python
#observation in the x axis it has taken classes of ages and in the y axis what is the frequency which falls in that particular
#classs
```

```python
#bins= devide your classes bydefault bins has been taken meaning the classes has been created in the x axis
#rwidth= our bars have occupied 100 percent width lets change the percentage of it 
#histType =  let us give step as value and see the darkness has been removed or the shade has been removed
#Allign = tells us where our bar should be middle left or right
#Orientation = we can make the bar left or right
##Label will not show untill we use the legend function
```

```python
bins = [15,20,25,30,35,40,45] #bins devide the classes
#plt.figure(figsize = (16,9))   


plt.hist(ml_students_age, bins, rwidth=0.8, histtype = "step",     #step no color inside
         orientation='vertical', color = "m", label = "ML Student")   #Label will not show untill we use the legend function

plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
#plt.legend()         #this legend depends on label in plot
plt.show()
```

```python

from matplotlib import style
style.use("ggplot")
bins = [15,20,25,30,35,40,45] #bins devide the classes
#plt.figure(figsize = (16,9))   


plt.hist(ml_students_age, bins, rwidth=0.8, histtype = "bar",  #hist type bar returns the bar
         orientation='horizontal', color = "c", label = "ML Student" )   #Label will not show untill we use the legend function

plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()         #this legend depends on label in plot
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
bins = [5,10,15,20,25,30,35,40,45] #bins devide the classes
#plt.figure(figsize = (16,9))   


plt.hist(ml_students_age, bins, rwidth=0.9, histtype = "bar",  #hist type bar returns the bar
         orientation='vertical', color = "c", label = "ML Student")   #Label will not show untill we use the legend function

plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend(loc=2)         #this legend depends on label in plot
plt.show()
```

```python

from matplotlib import style
style.use("ggplot")
bins = [10,15,20,25,30,35,40,45] #bins devide the classes
#plt.figure(figsize = (16,9))   


plt.hist(ml_students_age, bins, rwidth=0.8, histtype = "bar",align = "mid",  #hist type bar returns the bar
         orientation='vertical', color = "y", label = "ML Student")   #Label will not show untill we use the legend function

plt.title("ML Students age histograms",fontsize=15)
plt.xlabel("Students age cotegory",fontsize=13)
plt.ylabel("No. Students age",fontsize=13)
plt.legend()         #this legend depends on label in plot
plt.show()
```
